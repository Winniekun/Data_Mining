```python
import pyspark
import uuid
from pyspark import SQLContext
sc = pyspark.SparkContext('local','job_.{0}'.format(uuid.uuid4()))
sqlContext = SQLContext(sc)
```

```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv('../../data/PA_chapter2_data/movies.csv')
```

```python
rdd_data = sc.parallelize([list(r)[2:-1] for r in data.itertuples()])
```

```python
rdd_data.getNumPartitions()
```

```python
rdd_data.repartition(10).getNumPartitions()
```

```python
rdd_data.coalesce(2).getNumPartitions()
```

```python
import json
```

```python
rdd_data.map(lambda x: json.JSONEncoder().encode({str(k):str(v) for (k,v) in zip(data.columns[2:-1],x)})).saveAsTextFile('movie.json')
```

```python
df = sqlContext.read.json('movie.json')
```
